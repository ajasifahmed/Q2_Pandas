# Interacting with Databases

```python
import sqlite3
query = """
    CREATE TABLE test123
    (a VARCHAR(20), b VARCHAR(20),
    c REAL, d INTEGER);"""
print(query)
```

```python
con=sqlite3.connect('sdata.sqlite')
con.execute(query)
```

```python
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
('Tallahassee', 'Florida', 2.6, 3),
('Sacramento', 'California', 1.7, 5)]
```

```python
stmt = "INSERT INTO test123 VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
con.commit()
```

```python
cursor = con.execute('select * from test123')
rows = cursor.fetchall()
rows
```

```python
query= """
    CREATE TABLE test123
    (a VARCHAR(20), b VARCHAR(20),
    c REAL, d INTEGER);"""
con=sqlite3.connect('msdata.sqlite')
con.execute(query)
con.commit()
```

```python
data=[('Ali','Akaram',2.3,40),
       ('Alii','Akaraam',21.3,41),
       ('Ali','Akaram',2.3,40),
       ('Ali','Akaram',2.3,40)]
```

```python
r='INSERT INTO test123 VALUES(?,?,?,?)'
con.executemany(r,data)
con.commit()
cr=con.execute('select *from test123')
rows=cr.fetchall()
rows
```

```python
pd.DataFrame(rows,index=['a','b','c','d'],columns=['a','b','c','d'])
```

```python
import sqlalchemy as sq
d=sq.create_engine('sqlite:///msdata.sqlite')
l=pd.read_sql('select *from test123',d)
l
```
