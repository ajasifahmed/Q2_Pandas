```python
d=pd.DataFrame(np.arange(12).reshape(4,3),
               index=[['a', 'a', 'b', 'b'],[1,2,1,2]],
               columns=[['ohio','ohio','colarado'],['red','blue','green']]     
              )
d.index.names=['INDEX_1','INDEX_2']
d.columns.names=["STATE",'COLORS']
```

```python
d
```

```python
d.swaplevel(0,1)
#d.swaplevel('INDEX_1','INDEX_2')
```

```python
d.sort_index(level=1)
```

```python
print(d)

print(d.swaplevel(0,1).sort_index(level=0))

d.swaplevel(0,1).sort_index(level=1)
```

```python
display(d)
display(d.sum(level=0))#a+a and b+b
display(d.sum(level=1))#1+1 and 2+2
```

```python
display(d)
display(d.sum(level='STATE', axis=1))
display(d.sum(level='COLORS', axis=1))
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
display(df1)
display(df2)
```

```python
display(pd.merge(df1,df2))
display(pd.merge(df2,df1))
```

```python
display(pd.merge(df1,df2,on='key'))#good practice
print(pd.merge(df1,df2,how='outer'))
print(pd.merge(df1,df2,how='inner'))#same as the top one
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
display(df3)
display(df4)
```

```python
pd.merge(df3,df4, left_on='lkey', right_on='rkey')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],'key2': ['one', 'two', 'one'],'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'], 'key2': ['one', 'one', 'one', 'two'],'rval': [4, 5, 6, 7]})
display(left)
display(right)
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on='key1')
```

```python
data = pd.read_csv('macrodata.csv')
data.head()
```

```python
data.index = periods.to_timestamp('D', 'end')
display(periods)
data
```

```python
ldata = data.stack().reset_index().rename(columns={0: 'value'})
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```
